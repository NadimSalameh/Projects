# * Drop null values

```python
final_data.dropna(inplace = True)
```

```python
final_data.describe()
```

```python
final_data.info()
```

```python
final_data['population'] = final_data['population'].astype(int)
final_data['population']=final_data['population']/1000000
```

```python
import plotly.express as px
```

```python
condition =final_data['year']>=1960
```

```python
animated = final_data.loc[condition]
```

```python
animated.head()
```

# * Create an animated plot for life expectancy and fertility rate from 1960 till 2015

```python
px.scatter(animated, x="Life_expectancy", y="fertility_rate",animation_frame='year',animation_group="country",
           size='population',color='continent' ,hover_name="country", 
           size_max=45, range_x=[0,90], range_y=[0,10], title='Life expectancy - fertility rate from 1960 till 2015')
```
